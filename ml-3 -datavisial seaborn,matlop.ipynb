```python
import matplotlib.pyplot as plt
```

```python

cat=['Action','draw','thriller','comdey']
vote=[3,4,8,9]
plt.title("movie votecast")
plt.pie(vote,labels=cat,labeldistance=1.1)
plt.legend(loc=5)
plt.show()
```

```python
plt.pie(vote,labels=cat,labeldistance=0.5,explode=[0,0.5,0.2,0.1])
plt.legend(loc=4)
plt.show()
```

```python
plt.pie(vote,labels=cat,labeldistance=1.1,startangle=90)
plt.legend(loc=2)
plt.show
```

```python
plt.pie(vote,labels=cat,labeldistance=1.1,colors=['red','g','b','y'])
```

```python
plt.pie(vote,labels= cat,labeldistance=1.1,shadow=True,autopct="%0.2f%%",pctdistance=0.6)
```

```python
plt.pie(vote,labels=cat,labeldistance=1.1,radius=1)
```
